```python
import pandas as pd
import dask.dataframe as dd
import os
```

```python
cwd = os.getcwd()
data_path = os.path.join(cwd, 'DADOS','MICRODADOS_ENEM_2019.csv')
    
data = dd.read_csv(data_path, sep=';', encoding='ISO-8859-1')
data.head()
```

```python
results = []
for n in range(data.npartitions):
    df = data.partitions[n].compute()
    results.append(df.query('NU_IDADE in (13, 14)'))
    
alunos_menor_quatorze = pd.concat(results)
alunos_menor_quatorze.shape
```

```python
cwd = os.getcwd()
data_out = os.path.join(cwd, 'DADOS','OUT-IDADE-14-MICRODADOS_ENEM_2019.csv')
alunos_menor_quatorze.to_csv(data_out, index=False, encoding='UTF-8')
alunos_menor_quatorze.head()
```
